**Mini Proyek: Visualisasi Data Penjualan Kafe**

**Skenario:**

**Anda adalah seorang data scientist yang bekerja untuk sebuah jaringan kafe. Anda diberi tugas untuk membuat visualisasi yang membandingkan penjualan rata-rata harian antara tiga produk terlaris (Espresso, Latte, Cappuccino) selama seminggu. Visualisasi ini akan dipresentasikan kepada tim manajemen.**

**Tugas Anda:**

**Buatlah sebuah point plot menggunakan Seaborn yang menunjukkan tren penjualan rata-rata harian untuk ketiga produk tersebut. Anda harus melakukan kustomisasi pada plot agar terlihat profesional, informatif, dan mudah dibaca saat presentasi.**

----

**Dataset (Ilustrasi):**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh Data
data = {'hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'] * 3,
        'produk': ['Espresso']*7 + ['Latte']*7 + ['Cappuccino']*7,
        'penjualan_rata2': [65, 70, 80, 75, 90, 110, 120,
                            90, 95, 100, 110, 130, 140, 150,
                            85, 88, 95, 105, 120, 135, 145]}
df_penjualan = pd.DataFrame(data)

df_penjualan.head()
```

----

Kriteria Kustomisasi yang Harus Dipenuhi:

1. Gaya (Style): Gunakan gaya "whitegrid" untuk membantu perbandingan nilai pada sumbu y.

2. Skala (Context): Atur konteks menjadi "talk" agar semua elemen (teks, titik, garis) cukup besar untuk presentasi.

3. Palet Warna (Palette): Gunakan palet warna "Set2" atau palet kustom pilihan Anda agar setiap produk memiliki warna yang jelas dan menarik.

4. Plot Utama: Buat catplot dengan kind="point".

    - Sumbu X: hari

    - Sumbu Y: penjualan_rata2

    - Bedakan produk berdasarkan warna menggunakan argumen hue.

5. Judul dan Label:

    - Tambahkan judul utama yang informatif, misalnya: "Tren Penjualan Rata-Rata Harian Produk Kopi Unggulan". Posisikan judul agar tidak tumpang tindih dengan plot.

    - Ubah label sumbu x menjadi "Hari dalam Seminggu".

    - Ubah label sumbu y menjadi "Rata-Rata Penjualan (unit)".

6. Urutan Sumbu X: Pastikan hari pada sumbu x diurutkan dengan benar (Senin, Selasa, ..., Minggu).

7. Bonus: Coba gunakan palet warna kustom yang sesuai dengan citra merek kafe (misalnya, nuansa cokelat, krem, dan hijau).

---

```python
sns.set_style('whitegrid')
sns.set_context('talk')
sns.set_palette(['#8B4513'])
plot = sns.catplot(x='hari', y='penjualan_rata2', data=data, kind='point', height=6, aspect=16/6)
plot.fig.suptitle('Tren Penjualan Rata-Rata Harian Produk Kopi Unggulan')
plot.set(xlabel="Hari dalam Seminggu", ylabel="Rata-Rata Penjualan (Unit)")

plt.show()
```

**Hasil penjualan dari hari-kehari sejak dari hari senin semakin meningkat puncaknya ada di hari minggu**
**ini menunjukan minat customer terhadap kopi memang sangat tinggi tugas kita mempertahankan itu**
