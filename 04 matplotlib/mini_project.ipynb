**2. Distribusi Suhu (Histogram)**

```python
plt.figure(figsize=(6, 4))
plt.hist(suhu, bins=4)
plt.title("Distribusi Frekuensi Suhu")
plt.xlabel("Suhu (째C)")
plt.ylabel("Frekuensi Jumlah Bulan")


plt.show()

#hitung sebaran suhu
mean_suhu = np.mean(suhu)
median_suhu = np.median(suhu)
std_suhu = np.std(suhu)
print(f"Rata-rata Suhu: {mean_suhu:.2f} 째C")
print(f"Median Suhu: {median_suhu:.2f} 째C") 
print(f"Standar Deviasi Suhu: {std_suhu:.2f} 째C")
```

**3. Hubungan Suhu dan Curah Hujan (Scatter Plot)**

```python
plt.figure(figsize=(6, 4))
plt.scatter(suhu, curah_hujan)
plt.title("Hubungan Suhu dan Curah Hujan")
plt.xlabel("Suhu Rata-rata (째C)")
plt.ylabel("Curah Hujan (mm)")


plt.show()


#hitung korelasi suhu dan curah hujan
korelasi = np.corrcoef(suhu, curah_hujan)[0, 1]
print(f"Korelasi antara suhu dan curah hujan: {korelasi:.2f}")
```
