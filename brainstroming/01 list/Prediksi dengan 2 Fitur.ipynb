**Tantangan Studi Kasus 12: Prediksi dengan 2 Fitur**


```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    'jumlah_pengunjung': [50, 60, 80, 100, 120, 140],
    'suhu_celcius': [30, 32, 29, 26, 24, 22], # 30=Panas, 22=Sejuk
    'omzet_juta': [2.5, 3.1, 4.2, 5.5, 6.4, 7.5]
}

df = pd.DataFrame(data)

print("--- Data Historis (Pengunjung + Suhu) ---")
print(df)
```

- Memisahkan X dan y (Level Up!)

```python
# 1. Pisahkan Fitur (X)
# Kita pakai [[...]] (kurung siku ganda) agar hasilnya tetap DataFrame
X = df[['jumlah_pengunjung', 'suhu_celcius']]

# 2. Pisahkan Target (y)
y = df['omzet_juta']
```

**Latih dan Prediksi**

- Skenario Tes: Besok ramalan cuaca bilang suhu akan panas `(31Â°C)` dan kita menargetkan promosi besar-besaran agar datang `150` pengunjung. Berapa prediksi omzetnya?

```python
# 1. Latih Model
model = LinearRegression()
model.fit(X, y)


# Tampilkan Koefisien
print("Intercept:", model.intercept_)
print("Slope untuk jumlah_pengunjung:", model.coef_[0])
print("Slope untuk suhu_celcius:", model.coef_[1])

# 2. Siapkan Data Baru (Pakai DataFrame biar pro & aman dari warning)
# Ingat urutannya: [Pengunjung, Suhu]
data_baru = pd.DataFrame(
    [[150, 31]], 
    columns=['jumlah_pengunjung', 'suhu_celcius']
)

# 3. Prediksi
prediksi = model.predict(data_baru)

print(f"Prediksi Omzet: {prediksi[0]:.2f} Juta")
```

```python
import pandas as pd
import numpy as np

data = {
    'x': [50, 60, 80, 100, 120, 130],
    'y': [2.5, 3.0, 4.0, 5.0, 6.0, 6.5]
}
df = pd.DataFrame(data)

x = df['x']
y = df['y']

m_fast = ((x - x.mean()) * (y - y.mean())).sum() / ((x - x.mean())**2).sum()
c_fast = y.mean() - m_fast * x.mean()

print(f"fokus_slope",m_fast, f"fokus_intecept", c_fast)
```
