- Skenario: Bos Anda minta laporan: "Tolong buatkan tabel yang barisnya adalah Tanggal, kolomnya adalah Nama Produk, dan isinya adalah Jumlah Terjual."

----

```python
import pandas as pd

data = {
    'tanggal': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-01'],
    'produk': ['Roti', 'Susu', 'Roti', 'Susu', 'Kopi'],
    'terjual': [10, 5, 12, 8, 20]
}

df = pd.DataFrame(data)
print("--- Data Mentah ---")
print(df)
```

```python
pd.pivot_table(
    data=df, 
    index='tanggal',    # Apa yang mau jadi Baris?
    columns='produk',   # Apa yang mau jadi Kolom?
    values='terjual',   # Nilai apa yang mau dihitung?
    aggfunc='sum'       # Diapakan nilainya? (sum, mean, count)
)
```

```python
pivot_laporan = pd.pivot_table(
    data=df, 
    index='tanggal', 
    columns='produk', 
    values='terjual', 
    aggfunc='sum',
    fill_value=0  # <--- Jurus rahasia: Mengisi data kosong otomatis dengan 0
)

print("--- Laporan Pivot Final (Clean) ---")
print(pivot_laporan)
```
