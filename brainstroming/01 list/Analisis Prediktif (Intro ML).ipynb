**Menebak Omzet Kafe**

**Skenario: Bos kafe kamu bertanya, "Kita tahu omzet itu tergantung jumlah pengunjung. Bisa tidak kamu buatkan rumus sederhananya? Jadi kalau saya targetkan 100 pengunjung, kira-kira omzetnya berapa?"**

----

```python
import pandas as pd

data = {
    'jumlah_pengunjung': [50, 60, 80, 100, 120, 130],
    'total_omzet_juta': [2.5, 3.0, 4.0, 5.0, 6.0, 6.5]
}

df = pd.DataFrame(data)

print("--- Data Historis Kafe ---")
print(df)
```

**Tugas -> Mengajarkan "mesin" untuk mencari tahu hubungan antara `jumlah_pengunjung` dan `total_omzet_juta`**

```python
# 1. Pisahkan Fitur (X)
# Kita pakai [[...]] (kurung siku ganda) agar hasilnya tetap DataFrame
X = df[['jumlah_pengunjung']]

# 2. Pisahkan Target (y)
y = df['total_omzet_juta']

print("\n--- FITUR (SOAL) ---")
print(X)
print("\n--- TARGET (JAWABAN) ---")
print(y)
```

- Memanggil "Mesin" (Model)

```python
from sklearn.linear_model import LinearRegression

# 1. Buat "cetakan" modelnya
model = LinearRegression()

# 2. Latih modelnya! (Ini proses belajarnya)
model.fit(X, y)

print("--- MODEL SELESAI DILATIH! ---")
print("Model sudah belajar dari data X dan y.")
```

```python
# Melihat "Kemiringan" (m)
# Ini berarti: "Untuk setiap 1 pengunjung tambahan, omzet naik berapa juta?"
print(f"Koefisien/slope (m): {model.coef_[0]}")

# Melihat "Titik Awal" (c)
# Ini berarti: "Jika pengunjung 0, omzet (secara teori) mulai dari berapa juta?"
print(f"Intersep (c): {model.intercept_}")
```

```python
# Kita siapkan "soal" baru untuk ditebak model
# PENTING: Kita harus tetap pakai [[...]] (kurung siku ganda)
# karena modelnya mengharapkan data 2D (seperti tabel)
# Buat DataFrame dengan NAMA KOLOM yang sama persis
pengunjung_baru_df = pd.DataFrame(
    [[110]],  # <- Datanya
    columns=['jumlah_pengunjung'] # <- Nama fiturnya
)

# Suruh model menebak!
prediksi_omzet = model.predict(pengunjung_baru_df)

print(f"\n--- HASIL PREDIKSI ---")
# Cara yang benar untuk mengambil nilai dari DataFrame (baris ke-0, kolom ke-0)
print(f"Jika ada {pengunjung_baru_df.iloc[0, 0]} pengunjung, prediksi omzet adalah {prediksi_omzet[0]:.2f} juta Rupiah.")
```
