```python
# rata-rata pengunjung per jam
lambda_pengunjung = 15
```

**1. Simulasi Jumlah Kedatangan (Distribusi Poisson)**

- Gunakan fungsi poisson.rvs() untuk mensimulasikan jumlah kedatangan pengunjung selama 1000 jam. Gunakan lambda (Î») yang sesuai dengan informasi di atas.
- Buatlah histogram dari hasil simulasi Anda. Apa yang dapat Anda simpulkan dari bentuk distribusinya? Di manakah puncak distribusi tersebut?

```python
# menggunakan distribusi poisson untuk mensimulasikan jumlah kedatangan pelanggan
from scipy.stats import poisson
import matplotlib.pyplot as plt

# simulasi selama 1000 jam dengan rvs -> rvs ialah random variates sampling yang mana fungsi ini digunakan
# untuk menghasilkan data acak yang mengikuti distribusi poisson
size = 1000
data_pengunjung = poisson.rvs(lambda_pengunjung, size=size)
print(f"Jumlah Kemungkinan Pengunjung selama 1000 jam dengan rata-rata 5 pengunjung setiap hari:\n {data_pengunjung}")


# Buat histogram untuk menyimpulkan distribusi hasil simulasi
plt.figure(figsize=(16, 8))
plt.hist(data_pengunjung, bins=range(0, 40))

# Label dan judul
plt.title(f"Histogram Distribusi Poisson (Î»={lambda_pengunjung})")
plt.xlabel("Jumlah pengunjung per jam")
plt.ylabel("Frekuensi")
plt.show()


# additonal -> untuk mencari secara gamblang dan jelas kisaran mayoritas data(di sekitar mean) 
# dalam artian untuk mengatakan  âBiasanya pengunjung per jam antara X dan Y orang, dengan rata-rata Z.â
import numpy as np

# # Mean dan standard deviation (dari teori Poisson)
# mean = lambda_pengunjung
# std_dev = np.sqrt(lambda_pengunjung)

# # Interval berdasarkan aturan 68-95-99.7
# interval_1sigma = (mean - 1*std_dev, mean + 1*std_dev)
# interval_2sigma = (mean - 2*std_dev, mean + 2*std_dev)
# interval_3sigma = (mean - 3*std_dev, mean + 3*std_dev)

# print(f"Mean: {mean:.2f}")
# print(f"Standar deviasi: {round(std_dev)}")
# print(f"68% data (Â±1Ï): {tuple(map(round, interval_1sigma))}")
# print(f"95% data (Â±2Ï): {tuple(map(round, interval_2sigma))}")
# print(f"99.7% data (Â±3Ï): {tuple(map(round, interval_3sigma))}")


# data_pengunjung = poisson.rvs(mu=lambda_pengunjung, size=size)

# # Hitung batas mayoritas (95% data) dengan percentile
# low, high = np.percentile(data_pengunjung, [2.5, 97.5])

# print(f"Mayoritas (95%) pengunjung per hari ada di kisaran {int(low)} sampai {int(high)} orang")
# print(f"Rata-rata hasil simulasi: {np.mean(data_pengunjung):.2f}")


# --- Simulasi data ---
data_pengunjung = poisson.rvs(mu=lambda_pengunjung, size=size)

# --- Hasil teori (Î¼ Â± kÏ) menggunakan 95 karna kasus ini fokus pada mayoritas--- 
mean = lambda_pengunjung
std_dev = np.sqrt(lambda_pengunjung)
interval_teori = (mean - 2*std_dev, mean + 2*std_dev)

# --- Hasil simulasi (percentile) ---
low, high = np.percentile(data_pengunjung, [2.5, 97.5])

# --- Output ---
print("=== Hasil Teori (Î¼ Â± 2Ï) ===")
print(f"Rata-rata (Î¼): {mean}")
print(f"Standar deviasi (Ï): {std_dev:.2f}")
print(f"Kisaran mayoritas (95%): {tuple(map(round, interval_teori))}")

print("\n=== Hasil Simulasi (percentile) ===")
print(f"Rata-rata simulasi: {np.mean(data_pengunjung):.2f}")
print(f"Kisaran mayoritas (95%): ({int(low)}, {int(high)})")
```

Mayoritas data jatuh di kisaran 7â23 atau 8-23 pengunjung per jam, sesuai dengan nilai ekspektasi (mean = 15) dan standar deviasi â15 = 3.87 â 3.9 dari distribusi Poisson.â**

**2. Simulasi Waktu Antar Kedatangan (Distribusi Eksponensial)**

- Hitung nilai scale untuk distribusi eksponensial berdasarkan Î» dari proses Poisson (ingat, scale =1/Î»). Jangan lupa untuk mengubah unitnya menjadi menit!
- Gunakan expon.rvs() untuk mensimulasikan 1000 data "waktu tunggu" antar pengunjung dalam menit.
- Buatlah histogram dari data ini. Bagaimana bentuknya? Apakah simetris?

```python
# menghitung waktu antar kedatangan 
from scipy.stats import expon

#ubah unit ke satuan menit
lambda_per_menit = lambda_pengunjung / 60
# hitung scale
scale = 1 / lambda_per_menit

print(f"Î»/Rata-rata laju kedatangan (Î») per menit: {lambda_per_menit} pelanggan/menit")
print(f"Scale (rata-rata waktu antar kedatangan (scale = 1/Î»)) : {scale} menit")

print("=" * 50)

waktu_antar_kedatangan = expon.rvs(scale, size=size)
# print("Simulasi waktu tunggu (menit):\n", waktu_antar_kedatangan)
# Masukkan ke DataFrame
import pandas as pd
df = pd.DataFrame({"Waktu_Tunggu (menit)": waktu_antar_kedatangan})
print(df)


# Visualisasi distribusi
# Bins otomatis (sturges)
bins = int(np.log2(len(waktu_antar_kedatangan))) + 1

plt.hist(waktu_antar_kedatangan, bins=bins, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribusi Waktu Antar Kedatangan Pelanggan (Exponential)")
plt.xlabel("Waktu tunggu (menit)")
plt.ylabel("Frekuensi")
plt.show()

# melihat dengan cara menghitung apakah distribusi simetris atau skewed
from scipy.stats import skew
nilai_skew = skew(waktu_antar_kedatangan)
print(f"Skewness: {nilai_skew:.2f}")
```

**Berdasarkan simulasi distribusi eksponensial:**

- Rata-rata laju kedatangan (Î») per menit â 0.25 pelanggan/menit.
- Rata-rata waktu antar kedatangan (scale = 1/Î») â 4 menit.
- Histogram hasil simulasi menunjukkan bentuk right skewed, sesuai karakteristik distribusi eksponensial.
- Nilai skewness > 1.50, menegaskan bahwa distribusi ini tidak simetris.

**3. Buktikan Teorema Limit Pusat (CLT) dalam Aksi!**

- Dari data simulasi "waktu tunggu" (Distribusi Eksponensial) yang Anda buat di langkah 2, ambillah 1000 sampel acak, di mana masing-masing sampel berukuran 50.

- Hitung rata-rata dari setiap sampel tersebut. Anda sekarang akan memiliki 1000 nilai rata-rata.

- Buatlah histogram dari 1000 nilai rata-rata ini. Apa yang Anda amati? Bandingkan bentuk histogram ini dengan histogram di langkah 2 Sesuai dengan Teorema Limit Pusat, histogram ini seharusnya berbentuk seperti distribusi normal.

```python
# membuktikan dengan CLT bahwa bahwa rata-rata dari sampel acak yang cukup besar cenderung mengikuti distribusi normal 
# dengan mengulangi sampling sebanyak 1000 kali akan menghasilkan distribusi rata-rata sampel yang halus(mengambil sampel dari data eksponensial)

sample_means = []

for index in range(1000):
    sampel = np.random.choice(waktu_antar_kedatangan, size=50, replace=True)
    sample_means.append(np.mean(sampel))



# 3. Buat histogram
plt.figure(figsize=(12,5))

# Histogram data asli
plt.subplot(1,2,1)
plt.hist(waktu_antar_kedatangan, bins=50, color='skyblue', edgecolor='black')
plt.title("Data Awal (Eksponensial)")

# Histogram rata-rata sampel
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black')
plt.title("Distribusi Rata-rata (CLT)")

plt.show()
```

**CLT memungkinkan kita menggunakan distribusi normal meskipun data asli eksponensial.**

**4. Menjawab Pertanyaan Bisnis (menggunakan distribusi normal)**

- Dengan menggunakan distribusi normal dari rata-rata sampel yang Anda dapatkan di langkah 3, hitunglah:
    - Berapa probabilitas bahwa rata-rata waktu tunggu antar pengunjung adalah kurang dari 3 menit? (Gunakan norm.cdf()).
    - Berapa rata-rata waktu tunggu yang mewakili 90% pengunjung teratas (artinya, hanya 10% pengunjung yang menunggu lebih lama dari waktu ini)? (Gunakan norm.ppf()).

```python
from scipy.stats import norm

# menghitung probabilitas waktu tunggu pengunjung kurang dari 3 menit -> persentase/probabilitas
mean_sample = np.mean(sample_means)
std_sample = np.std(sample_means)
# Gunakan scipy.stats.norm.cdf(x, loc, scale) 
prob_less_3 = norm.cdf(3, loc=mean_sample, scale=std_sample)
print(f"Probabilitas rata-rata waktu tunggu Pengunjung < 3 menit: {prob_less_3} hampir 0%,\nartinya hampir tidak mungkin rata-rata pengunjung menunggu kurang dari 3 menit.")

print("=" * 50)

# Rata-rata yang mewakili 90% pengunjung teratas
# Artinya mencari quantile 90%, pakai norm.ppf() -> top 10 persen
top_10_percent = norm.ppf(0.9, loc=mean_sample, scale=std_sample)
print(f"Rata-rata untuk 10% pengunjung menunggu paling lama: {top_10_percent} menit")
```

- **kebanyakan pengunjung menunggu lebih lama dari 3 menit.**
- **10 persen pengunjung menunggu paling lama sebesar 5.12 menit**
